## Numpy array and basics

```python
import numpy as numpy
import time
```

Creating Array from List

```python
arr_1d = numpy.array([1, 2, 3, 4, 5])
print(arr_1d)

arr_2d = numpy.array([[1, 2, 3], [4, 5, 6]]) #numpy.array takes a single array
print(arr_2d)
```

### List vs NumPy Array

```python
py_list = [1, 2, 3]
print("Python list multiplication ", py_list * 2)

np_array = numpy.array([1, 2, 3]) #element wise multiplication
print("Python array multiplication ", np_array * 2)
```

```python
start = time.time()
py_list = [i*2 for i in range(10000000)]
print("\n List operation time: ", time.time() - start)
```
