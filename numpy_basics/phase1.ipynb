## Numpy array and basics

```python
import numpy as numpy
import time
```

Creating Array from List

```python
arr_1d = numpy.array([1, 2, 3, 4, 5])
print(arr_1d)

arr_2d = numpy.array([[1, 2, 3], [4, 5, 6]]) #numpy.array takes a single array
print(arr_2d)
```

### List vs NumPy Array

```python
py_list = [1, 2, 3]
print("Python list multiplication ", py_list * 2)

np_array = numpy.array([1, 2, 3]) #element wise multiplication
print("Python array multiplication ", np_array * 2)
```

```python
start = time.time()
py_list = [i*2 for i in range(10000000)]
print("\n List operation time: ", time.time() - start)

start = time.time()
np_array = numpy.arange(100000) * 2
print("\n Numpy operation time: ", time.time() - start)
```

### Creating Array from Scratch

```python
zeros = numpy.zeros((3, 4))
print("zeros array: \n", zeros)

ones = numpy.ones((2, 3))
print("Ones Array: \n", ones)

full = numpy.full((2, 2), 7)
print("Full Array: \n", full)
```

```python
random = numpy.random.random((2, 3))
print("Random Array: \n", random) #Fluctuates between 0s & 1s

sequence = numpy.arange(0, 10, 2)
print("Sequence Array: \n", sequence)
```

### Vector, Matrices, & Tensor

```python
vector = numpy.array([1, 2, 3])
print("Vector: ", vector)

matrix = numpy.array([[1, 2, 3],
                      [4, 5, 6]])
print("Matrix: \n", matrix)

tensor = numpy.array([[[1, 2] , [3, 4]],
                      [[5, 6], [7, 8]]])
print("Tensor: \n", tensor)
```

### Array Properties 

```python
arr = numpy.array([[1, 2, 3],
                   [4, 5, 6]])
print("Shape", arr.shape)
print("Dimension", arr.ndim)
print("Size", arr.size)
print("Data Type", arr.dtype)
```

### Array Reshaping

```python
arr = numpy.arange(12)
print("Orignal array: ", arr)

reshaped = arr.reshape((3, 4))
print("\n Reshaped array: \n", reshaped)

flattened = reshaped.flatten()
print("\n Flattened array: ", flattened)

# returns view, instead of copy
raveled = reshaped.ravel()
print("\n Raveled array: ", raveled)

# Transpose
transpose = reshaped.T
print("\n Transposed array: \n", transpose)
```
